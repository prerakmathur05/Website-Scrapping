```python
# The Website used here is designed to be scraped so copyright issues are not there!
import requests,bs4

url="http://books.toscrape.com/catalogue/page-{}.html"
pages=[n+1 for n in range(10)]
#r=requests.get(url.format(1))

mybooks=[]
for n in pages:
    r=requests.get(url.format(n))
    soup=bs4.BeautifulSoup(r.text,'lxml')
    product=soup.select('.product_pod')
    for p in product:
        if (p.select('.star-rating.Three')==[] and p.select('.star-rating.Four')==[] and p.select('.star-rating.Five')==[]):
            pass
        else:
            book=p.select('a')[1]['title']
            mybooks.append(book)

    
    
    
    
print(f'Yours book collection ready, You have {len(mybooks)} books in total')
for b in mybooks:
    print(b)
    
    
    
```
